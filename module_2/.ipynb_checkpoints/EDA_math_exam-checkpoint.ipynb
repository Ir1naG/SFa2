```python
#function for the general check of the data correctness inside columns

def general_col_check(col):
    col = col.astype(str).apply(lambda x: None if x.strip()=='' else x)
    print("Amount of met values:\n", col.value_counts())
    print('\nTop 10 most frequently met values:', (col.value_counts()>10).sum())
    print('\nAmount of unique values {}'.format(col.nunique()))
    print('\nAmount of zero values {}'.format(col.isnull().sum()))
    print('\nInformation about the column:')
    pd.DataFrame(col).info()
    
    return
```

```python
def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=students.loc[students.loc[:, column].isin(students.loc[:, column].value_counts().index[:10])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

Сильно скоррелированы между собой переменные Medu и Fedu, при этом с переменной Score образование матери скоррелировано больше.

Из числовых признаков сильнее скоррелированы с переменной scpre следующие:
Age (обратно), Medu, Fedu (чем лучше образование родителей, тем лучше оценка), Studytime (логично), Failures (обратная, чем меньше, тем лучше оценка), goout (обратная)
Интересно, что количество пропусков никак не скоррелировано

```python
for col in ['school', 'sex', 'address', 'Pstatus', 'Mjob', 'Fjob', 'schoolsup',
            'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet',
            'romantic']:
    get_boxplot(col)
```

```python
def get_stat_dif(column):
    cols = students.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(students.loc[students.loc[:, column] == comb[0], 'score'], 
                     students.loc[students.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
for col in ['school', 'sex', 'address', 'Pstatus', 'Mjob', 'Fjob', 'schoolsup',
            'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet',
            'romantic']:
    get_stat_dif(col)
```
